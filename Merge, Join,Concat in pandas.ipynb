```python
import pandas as pd
```

```python
df= pd.read_excel(r'Documents/Excel functions.xlsx', sheet_name='Substitute')
df
```

```python
#merge- natural(merges similar df),full(merges everything),leftouter(merges all in left df and similar in right),right(everythin in right df and similar)
```

```python
df2= pd.read_excel(r'Documents/Excel functions.xlsx', sheet_name='SUM-SumIF')
df2
```

```python
df.merge(df2, how='inner',on='EmployeeID')
```

```python
df.merge(df2, how='inner',on=['FirstName','EmployeeID'])
```

```python
df.merge(df2, how='outer',on='EmployeeID')
```

```python
df.merge(df2, how='left',on='EmployeeID')
```

```python
df.merge(df2, how='right',on='EmployeeID')
```

```python
df.merge(df2, how='cross')

 #cross joins-compare each value to a list in the whole table  then to nextvalue )
```

```python
df.join(df2, on='EmployeeID',how='outer',lsuffix ='_Left', rsuffix ='_Right')
```

```python
df3 =df.set_index('EmployeeID').join(df2.set_index('EmployeeID'), lsuffix ='_Left', rsuffix ='_Right')
df3

#df3 =df.set_index('EmployeeID').join(df2.set_index('EmployeeID'), lsuffix ='_Left', rsuffix ='_Right' how="(use the join you want))
df3
```
